# Functions to acquire data

## 2020 Census Function

```python
# going for the 2020 deciennial census data instead of the american community sruvey becasue the zcta is an area that we can actually pull

def get_2020_census_data():
    '''
    This function gets 2020 decienn=ial data from the census bureau api website for all Zip Code Tabulation Areas (ZCTAs) in the state of Indiana
    Modules:
        import pandas as pd
        import requests
    '''
    # assign url
    url = 'https://api.census.gov/data/2020/dec/dp?get=group(DP1)&for=zip%20code%20tabulation%20area%20(or%20part):*&in=state:18'

    # get data and turn into json
    r = requests.get(url).json()

    # create data frame
    df = pd.DataFrame(r[1:], columns=r[0])

    return df
```

## 2020 Census Labels Function

```python
def get_2020_census_labels():
    '''
    This function pulls that labels associated woith th column names for the 2020 diciennial census data
    '''
    # assign url
    url = 'https://api.census.gov/data/2020/dec/dp/groups/DP1/'

    # get data and turn into json
    r = requests.get(url).json()

    # convert to df
    df = pd.DataFrame(r['variables']).T

    return df
```
